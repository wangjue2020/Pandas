## 1. 经过统计得到多维度指标数据

非常常见的统计场景，指定多个唯独，计算聚合后的指标  
实例：统计得到电影评分数据集，每个月份的每个分数被评分多少次：（月份，分数1-5， 次数）

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('./datas/movielens-1m/ratings.dat',
              header=None,
              names=['userid','movieid','rating','timestamp'],
              sep='::',
              engine='python')
df.head()
```

```python
df['pdate']=pd.to_datetime(df['timestamp'],unit='s')
df.head()
```

```python
df.dtypes
```

```python
#实现数据统计
df_group=df.groupby([df['pdate'].dt.month,'rating'])['userid'].agg(pv=np.sum)
df_group.head(20)
```

对这样格式的数据，我们想查看按月份，不同评分的次数趋势，是没法实现的  
需要将数据变成每个评分是一列才行

## 2. 使用unstack实现数据二维透视  
目的：想要画图对比按照月份的不同的评分的数量趋势

```python
df_stack = df_group.unstack()
df_stack
```

```python
df_stack.plot()
```

```python
df_stack.stack().head(20)
```

```python
df_reset = df_group.reset_index()
df_reset.head()
```

```python
df_pivot=df_reset.pivot('pdate','rating','pv')
df_pivot.head()
```

```python
df_pivot.plot()
```

pivot 方法相当于对df使用set_index创建分层索引，然后调用unstack

## 4. stack、unstack、pivot语法  

stack: DataFrame.stack(level=1, dropna=True) 将column变成index  

level=-1代表多层索引的最内层，可以通过0，1，2指定多层索引的对应层

pivot: DataFrame.pivot(index=None,columns=None,values=None), 指定index，columns，values实现二维透视
